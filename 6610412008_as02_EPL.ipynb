```python
import pandas as pd
import numpy as np
import sklearn as sk
```

```python
file_path = "https://raw.githubusercontent.com/jakphunn/test/main/2021-2022.csv"
df = pd.read_csv(file_path)
df.head(10)
```

```python
#Q1 show the top 3 teams that have the highest scores.

score = pd.DataFrame()

score['Team'] = pd.concat([df['HomeTeam'], df['AwayTeam']])
score['Goals'] = pd.concat([df['FTHG'], df['FTAG']])
total_goal = score.groupby('Team')['Goals'].sum().reset_index()
top3_team = total_goal.sort_values(by='Goals', ascending=False).head(3)

print("Top Three Teams with Highest Total Goals:")
print(top3_team.to_string(index=False))
```

```python
#Q2 show the best home and away teams at shooting on target.

df['HSP'] = df['HST'] / df['HS']
df['ASP'] = df['AST'] / df['AS']

ht = df.loc[df['HSP'].idxmax()]['HomeTeam']
at = df.loc[df['ASP'].idxmax()]['AwayTeam']

print(f'Team with the highest shot precision as Home Team: {ht}')
print(f'Team with the highest shot precision as Away Team: {at}')
```

```python
#Q3 how many times that Manchester United and Liverpool match each other in 2021-2022

lp = df[(df['HomeTeam'] == 'Liverpool') | (df['AwayTeam'] == 'Liverpool')]

mu = df[(df['HomeTeam'] == 'Man United') | (df['AwayTeam'] == 'Man United')]

lvsm = pd.concat([lp, mu])

print("Number of matches between Liverpool and Manchester United:", len(lvsm))
```

```python
#Q4 the results of Liverpool vs Man utd in 2021-2022

lp = df[(df['HomeTeam'] == 'Liverpool') & (df['AwayTeam'] == 'Man United')]

mu = df[(df['HomeTeam'] == 'Man United') & (df['AwayTeam'] == 'Liverpool')]

lvsm = pd.concat([lp, mu])

lp_stats = {
    'Wins': lvsm[lvsm['FTR'] == 'H'].shape[0],
    'Draws': lvsm[lvsm['FTR'] == 'D'].shape[0],
    'Losses': lvsm[lvsm['FTR'] == 'A'].shape[0]
}

mu_stats = {
    'Wins': lvsm[lvsm['FTR'] == 'A'].shape[0],
    'Draws': lvsm[lvsm['FTR'] == 'D'].shape[0],
    'Losses': lvsm[lvsm['FTR'] == 'H'].shape[0]
}

print("Liverpool Statistics:")
print(lp_stats)

print("\nManchester United Statistics:")
print(mu_stats)
```

```python
#Q5 the top 3 teams that have the highest comeback rates in 2021-2022

cb_home = (df['HTR'] != 'H') & (df['FTR'] == 'H')
cb_away = (df['HTR'] == 'H') & (df['FTR'] == 'A')
cb_match = df[cb_home | cb_away]

cb_team = pd.concat([cb_match['HomeTeam'], cb_match['AwayTeam']])

cb_count = cb_team.value_counts()

total_match = pd.concat([df['HomeTeam'], df['AwayTeam']]).value_counts()
cb_rate = cb_count / total_match

top3_cb_team = cb_rate.nlargest(3)

print("Top 3 Teams with the Highest Comeback Rates:")
for team, rate in top3_cb_team.items():
    print(f"{team}: {rate:.2%}")
```
